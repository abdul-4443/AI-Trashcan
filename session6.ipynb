```python
import math

leaves = [3, 5, 6, 9, 1, 2, 0, -1]
pruned = []

def ab(idx, depth, is_max, a, b):
    if depth == 2:
        return leaves[idx]

    if is_max:
        best = -math.inf
        print(f"[MAX] a={a}  b={b}")
        for i in range(4):         
            v = ab(i * 2, depth+1, False, a, b)
            best = max(best, v)
            a = max(a, best)
            print(f"  MIN child {i} => {v}  a={a}")
            if b <= a:
                pruned.append(f"MAX level: pruned after child {i}")
                print(f"  PRUNED rest at MAX level")
                break
        return best
    else:
        best = math.inf
        print(f"  [MIN {idx//2}] a={a}  b={b}")
        for i in range(2):
            v = ab(idx + i, depth+1, False, a, b)
            print(f"    leaf {idx+i} = {v}")
            best = min(best, v)
            b = min(b, best)
            print(f"    b={b}")
            if b <= a:
                pruned.append(f"MIN {idx//2}: pruned after leaf {idx+i}")
                print(f"    PRUNED rest at MIN {idx//2}")
                break
        return best

print("=" * 40)
opt = ab(0, 0, True, -math.inf, math.inf)
print("=" * 40)
print(f"Optimal root value: {opt}")
print(f"Pruned branches:")
for p in pruned:
    print(f"  - {p}")
```

```python
def elev(x):
    return -(x - 3)**2 + 9

def climb(start=0, lo=0, hi=6):
    cur = start
    path = [cur]
    print(f"{'Pos':>6} {'Elev':>8} {'Move'}")
    print("-" * 30)
    while True:
        nb = [x for x in [cur-1, cur+1] if lo <= x <= hi]
        best = max(nb, key=elev)
        print(f"{cur:>6} {elev(cur):>8.1f}", end="")
        if elev(best) > elev(cur):
            print(f"  -> x={best}")
            cur = best
            path.append(cur)
        else:
            print(f"  STOP")
            break
    print(f"\nPath   : {path}")
    print(f"Final  : x={cur}  elev={elev(cur)}")

climb(0)
```

```python
g = {
    'A': ['B','C'],
    'B': ['D'],
    'C': ['D'],
    'D': ['E','F'],
    'E': [],
    'F': []
}

score = {'A':3,'B':6,'C':5,'D':7,'E':9,'F':8}

def hc(start='A'):
    cur = start
    path = [cur]
    print(f"{'State':>8} {'Score':>8} {'Neighbors'}")
    print("-" * 35)
    while True:
        nb = g[cur]
        print(f"{cur:>8} {score[cur]:>8}  {nb}")
        if not nb:
            print("  No neighbors — stopped.")
            break
        nxt = max(nb, key=lambda s: score[s])
        if score[nxt] > score[cur]:
            cur = nxt
            path.append(cur)
        else:
            print("  All neighbors worse — LOCAL MAX, stopped.")
            break
    best = max(score, key=score.get)
    print(f"\nPath         : {' -> '.join(path)}")
    print(f"Final state  : {cur}  score={score[cur]}")
    print(f"Global opt   : {best}  score={score[best]}")
    if cur != best:
        print("Stuck at local max — missed global optimum.")
    else:
        print("Found global optimum!")

hc('A')
```

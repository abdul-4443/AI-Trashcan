```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

np.random.seed(99)
n = 40

mail_len       = np.random.randint(50, 2000, n)
link_count     = np.random.randint(0, 20, n)
special_chars  = np.random.randint(0, 50, n)
has_free       = np.random.randint(0, 2, n)

spam_prob = (
    0.003 * mail_len +
    0.15  * link_count +
    0.08  * special_chars +
    0.4   * has_free
)
is_spam = (spam_prob > spam_prob.mean()).astype(int)

mailbox = pd.DataFrame({
    "mail_len":mail_len,
    "link_count":link_count,
    "special_chars":special_chars,
    "has_free":has_free,
    "is_spam":is_spam
})

features = mailbox.drop("is_spam", axis=1)
labels   = mailbox["is_spam"]

Xtr, Xte, ytr, yte = train_test_split(features, labels, test_size=0.2, random_state=3)

rescaler  = MinMaxScaler()
Xtr_norm  = rescaler.fit_transform(Xtr)
Xte_norm  = rescaler.transform(Xte)

proximity_clf = KNeighborsClassifier(n_neighbors=5, metric="euclidean")
proximity_clf.fit(Xtr_norm, ytr)
knn_preds = proximity_clf.predict(Xte_norm)

logit_clf = LogisticRegression(max_iter=500)
logit_clf.fit(Xtr_norm, ytr)
logit_preds = logit_clf.predict(Xte_norm)

print(f"KNN Accuracy: {accuracy_score(yte, knn_preds):.4f}")
print(f"LogReg Accuracy: {accuracy_score(yte, logit_preds):.4f}")

winner = "KNN" if accuracy_score(yte, knn_preds) > accuracy_score(yte, logit_preds) else "Logistic Regression"
print(f"\nBetter Model: {winner}")
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

np.random.seed(11)
n = 28

sq_footage   = np.random.randint(800, 4500, n)
room_count   = np.random.randint(2, 8, n)
locale_score = np.round(np.random.uniform(1, 10, n), 1)
house_age    = np.random.randint(1, 50, n)
noise        = np.random.normal(0, 5000, n)

valuation = (
    120 * sq_footage +
    8000 * room_count +
    15000 * locale_score -
    500 * house_age +
    noise
)

estate_df = pd.DataFrame({
    "sq_footage":sq_footage,
    "room_count":room_count,
    "locale_score":locale_score,
    "house_age":house_age,
    "valuation":valuation
})

attrs  = estate_df.drop("valuation", axis=1)
prices = estate_df["valuation"]

Xtr, Xte, ytr, yte = train_test_split(attrs, prices, test_size=0.2, random_state=5)

zscore = StandardScaler()
Xtr_z  = zscore.fit_transform(Xtr)
Xte_z  = zscore.transform(Xte)

regessor = LinearRegression()
regessor.fit(Xtr_z, ytr)

projections = regessor.predict(Xte_z)

mse  = mean_squared_error(yte, projections)
rmse = np.sqrt(mse)
r2   = r2_score(yte, projections)

print(f"MSE: {mse:,.2f}")
print(f"RMSE: {rmse:,.2f}")
print(f"R^2: {r2:.4f}")
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score

np.random.seed(77)
n = 35

exp_years  = np.random.randint(0, 30, n)
edu_tier   = np.random.randint(1, 5, n)
aptitude   = np.round(np.random.uniform(40, 100, n), 1)
age        = 22 + exp_years + np.random.randint(0, 5, n)
disturbance= np.random.normal(0, 3000, n)

compensation = (
    2500 * exp_years +
    8000 * edu_tier +
    300  * aptitude +
    500  * age +
    disturbance
)

workforce = pd.DataFrame({
    "exp_years": exp_years,
    "edu_tier": edu_tier,
    "aptitude": aptitude,
    "age": age,
    "compensation": compensation
})

covariates = workforce.drop("compensation", axis=1)
remuneration = workforce["compensation"]

Xtr, Xte, ytr, yte = train_test_split(covariates, remuneration, test_size=0.2, random_state=13)

ols_model = LinearRegression()
ols_model.fit(Xtr, ytr)
ols_hat = ols_model.predict(Xte)

dt_model = DecisionTreeRegressor(max_depth=4, random_state=13)
dt_model.fit(Xtr, ytr)
dt_hat = dt_model.predict(Xte)

ols_r2 = r2_score(yte, ols_hat)
dt_r2  = r2_score(yte, dt_hat)

print(f"Linear Regression R^2: {ols_r2:.4f}")
print(f"Decision Tree R^2: {dt_r2:.4f}")
print(f"\nWinner: {'Linear Regression' if ols_r2 > dt_r2 else 'Decision Tree'}")
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

np.random.seed(55)
n = 40

patron_age    = np.random.randint(18, 65, n)
wallet_size   = np.random.randint(20000, 150000, n)
prior_buys    = np.random.randint(0, 20, n)
dwell_minutes = np.round(np.random.uniform(0.5, 60, n), 1)

tendency = (
    0.3  * patron_age +
    0.00005 * wallet_size +
    1.5  * prior_buys +
    0.08 * dwell_minutes
)
purchased = (tendency > tendency.mean()).astype(int)

storefront = pd.DataFrame({
    "patron_age":patron_age,
    "wallet_size":wallet_size,
    "prior_buys":prior_buys,
    "dwell_minutes":dwell_minutes,
    "purchased":purchased
})

signals = storefront.drop("purchased", axis=1)
verdict = storefront["purchased"]

Xtr, Xte, ytr, yte = train_test_split(signals, verdict, test_size=0.2, random_state=21)

centering = StandardScaler()
Xtr_c = centering.fit_transform(Xtr)
Xte_c = centering.transform(Xte)

lm_pipe = LogisticRegression(max_iter=400)
lm_pipe.fit(Xtr_c, ytr)
lm_calls = lm_pipe.predict(Xte_c)

knn_pipe = KNeighborsClassifier(n_neighbors=7)
knn_pipe.fit(Xtr_c, ytr)
knn_calls = knn_pipe.predict(Xte_c)

print(f"LogReg Accuracy: {accuracy_score(yte, lm_calls):.4f}")
print(f"KNN Accuracy: {accuracy_score(yte, knn_calls):.4f}")

prospect = pd.DataFrame([[28, 55000, 4, 22.5]], columns=signals.columns)
prospect_sc = centering.transform(prospect)

print("\nNew Customer Prediction:")
print(f"LogReg: {'Buy' if lm_pipe.predict(prospect_sc)[0] == 1 else 'No Buy'}")
print(f"KNN: {'Buy' if knn_pipe.predict(prospect_sc)[0] == 1 else 'No Buy'}")
```
